```python
# Assuming df1 is your DataFrame

# Combine 'title' with 'employment_types' for normalization
combined_data = [{'title': row['title'], 'employment_types': row['employment_types']} for _, row in df3.iterrows()]

# Normalize the combined data
employment_types_normalized3 = pd.json_normalize(
    combined_data,
    record_path='employment_types',
    meta='title',
    sep='_',
    errors='ignore'
)

# Print the name of columns
print("Columns in employment_types_normalized:")
print(employment_types_normalized3.columns)
```

```python
import ast

# Convert the string representation of the list to an actual list
df3['skills'] = df3['skills'].apply(ast.literal_eval)

# Combine 'title' with 'skills' for normalization
combined_data_skills = [{'title': row['title'], 'skills': row['skills']} for _, row in df3.iterrows()]

# Normalize the combined skills data
skills_normalized = pd.json_normalize(
    combined_data_skills,
    record_path='skills',
    meta='title',
    sep='_',
    errors='ignore'
)

# Print the name of columns in skills_normalized
print("Columns in skills_normalized:")
print(skills_normalized.columns)
skills_normalized.head(20)
```

```python
import pandas as pd
import ast

# Assuming df2 is your DataFrame

# Check if 'multilocation' is not NaN and is a string before applying literal_eval
def safe_literal_eval(x):
    try:
        return ast.literal_eval(x) if pd.notna(x) and isinstance(x, str) else None
    except (SyntaxError, ValueError):
        return None

df3['multilocation'] = df3['multilocation'].apply(safe_literal_eval)

# Combine 'title' with 'multilocation' for normalization
combined_data_multilocation = [{'title': row['title'], 'multilocation': row['multilocation']} for _, row in df3.iterrows()]

# Normalize the combined multilocation data
multilocation_normalized2 = pd.json_normalize(
    combined_data_multilocation,
    record_path='multilocation',
    meta='title',
    sep='_',
    errors='ignore'
)

# Print the name of columns in multilocation_normalized
print("\nColumns in multilocation_normalized:")
print(multilocation_normalized2.columns)
```

```python
# Assuming employment_types_normalized is the normalized DataFrame

# Specify the desired columns
selected_columns = ['city', 'street', 'slug', 'title']

# Create the new table with selected columns
multilocation_normalized2 = multilocation_normalized2[selected_columns]

# Display the new table
multilocation_normalized2.head(10)
```
